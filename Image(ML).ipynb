### Basic methods

### 1.How to convert an 2D Image into machine representation value i.e array
cv2.imread(path of 2d image,flags = 0(gray scale colour space)) defaultly it will be--->3D array--->(BGR) it can be a jpg,jpeg,png.,but when we use flags = 0 it will be converted to 2D array.

### cv2.imread(): 
To convert 2d image to a array.
### cv2.imshow():
To display a array into a image.
### cv2.waitkey():
To add a delay i.e how much milliseconds the window will be activated in the screen.
### cv2.destroyAllWindow(): 
To distroy the window
### cv2.merge():
To merge the image colour space
### cv2.resize():
To resize the array into required format
### cv2.imwrite():
To convert array into image
